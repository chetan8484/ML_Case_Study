```python
# Categorizing the data as independent and dependent columns 

x = df.drop(['Converted'] , axis=1).values
y = df['Converted'].values
```

```python
# Standardizing the data using StandardScaler Method

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
df.head()
```

```python
# Spliting the data into train and test Data.
from sklearn.model_selection import train_test_split

x_train , x_test, y_train , y_test = train_test_split(x,y , test_size=0.2,random_state = 0)
```

# Creating Model:Logistic Regression.

```python
# Creating a model
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()

lr.fit(x_train , y_train)
```

```python
y_pred = lr.predict_proba(x_test) 
```

```python
y_pred
```

```python
lr.classes_
```

```python
pred_prob_list = []
for i in lr.predict_proba(x):
    pred_prob_list.append(round(i[1]*100,2))
    
pred_prob_list
```

```python
d1 = {'Lead Number' : df_original['Lead Number'].tolist(), 'Leads Score' : pred_prob_list}

pd.DataFrame(d1)

#a lead score between 0 and 100 to each of the leads which can be used by the company to target potential leads.
#A higher score would mean that the lead is hot, i.e. is most likely to convert whereas a lower score would mean that 
#the lead is cold and will mostly not get converted.
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,lr.predict(x_test)))
```

```python
# Displaying True-Postive and True-Negative with use of Confusion Matrix.
from sklearn.metrics import confusion_matrix , accuracy_score

cm = confusion_matrix(lr.predict(x_test) , y_test)
```

```python
cm
```

```python
accuracy_score(lr.predict(x_test) , y_test)
```
